```python
def initialize(dimension):
    
    weigth = np.full((dimension,1),0.01)
    bias = 0.0
    return weigth,bias
```

```python
def sigmoid(z):
    y_head=1/(1+np.exp(-z))
    return y_head
```

```python
def forwardBackward(weigth,bias,x_train,y_train):
    #Forward
    y_head=sigmoid(np.dot(weigth.T,x_train)+bias)
    loss=-(y_train*np.log(y_head)+(1-y_train)*np.log(1-y_head))
    cost=np.sum(loss)/x_train.shape[1]
    
    
    #Backward
    derivative_weigth = np.dot(x_train,((y_head-y_train).T))/x_train.shape[1]
    derivative_bias = np.sum(y_head-y_train)/x_train.shape[1]
    gradients = {"Derivative Weight" : derivative_weigth, "Derivative Bias" : derivative_bias}
    
    return cost,gradients
    
```

```python
def update(weigth,bias,x_train,y_train,learningRate,iteration):
    costlist=[]
    indexs=[]
    for i in range(iteration):
        cost,gradient=forwardBackward(weigth,bias,x_train,y_train)
        weigth=weigth-learningRate*gradient['Derivative Weight']
        bias=bias-learningRate*gradient['Derivative Bias']
        costlist.append(cost)
        indexs.append(i)
        
        
    parameters = {"weight": weigth,"bias": bias}
    
    print("iteration:",iteration)
    print("cost:",cost)

    plt.plot(indexs,costlist)
    plt.xlabel("Number of Iteration")
    plt.ylabel("Cost")
    plt.show()

    return parameters, gradient
        
```

```python
def predict(weigth,bias,x_test):
    
    #print(weigth.shape)
    #print(x_test.shape)
    
    z=np.dot(weigth.T,x_test)+bias
    y_head=sigmoid(z)
    print(y_head)
    y_prediction=np.zeros((1,x_test.shape[1]))
    print(y_prediction)
    for i in range(y_head.shape[1]):
        if (y_head[0,i]<=0.5):
            y_prediction[0,i]=0
        else:
            y_prediction[0,i]=1
    return y_prediction
```

```python
def logistic_regression(x_train,y_train,x_test,y_test,learningRate,iteration):
    dimension = x_train.shape[0]
    weight,bias = initialize(dimension)
    
    parameters, gradients = update(weight,bias,x_train,y_train,learningRate,iteration)
    print(parameters['weight'].shape)
    print(x_train.shape)
    print('bias')
    print(parameters['bias'])
    y_prediction = predict(parameters["weight"],parameters["bias"],x_test)
    
    print(y_prediction)
    print("Manuel Test Accuracy: {:.2f}%".format((100 - np.mean(np.abs(y_prediction - y_test))*100)))
```
